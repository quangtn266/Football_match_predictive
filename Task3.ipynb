<h2>Data preprocess</h2>

Based on features such as "home_team_coach_id" from "train.csv" and "score" from "train_target_and_scores.csv"

```python
update_df_train = pd.concat([df_train_ts['score'], df_train],axis=1)
update_df_train.head()
```

```python
missval_cols= update_df_train.columns[update_df_train.isnull().any()].tolist()
for i in range(len(missval_cols)):
    update_df_train_ = update_df_train.drop(update_df_train.index
                                [update_df_train[update_df_train[missval_cols[i]].isnull()].index.tolist()])
update_df_train_.head()
```

```python
missval_cols= df_test.columns[df_test.isnull().any()].tolist()
for i in range(len(missval_cols)):
    update_df_test = df_test.drop(df_test.index[df_test[df_test[missval_cols[i]].isnull()].index.tolist()])
update_df_test.head()
```

```python
update_df_train_drop = update_df_train.copy()
update_df_train_drop.drop(['home_team_name', 'away_team_name','target','score','id'], axis='columns',inplace=True)
update_df_train_drop.head()
```

```python
update_df_test_ = df_test.copy()
update_df_test_.drop(['home_team_name', 'away_team_name','id'], axis='columns',inplace=True)
update_df_test_.head()
```

```python
df_train_ts.head()
```

```python
X = update_df_train_drop.copy()
y = pd.DataFrame(update_df_train.score)
```

<h3>Handling Categorical values</h3>

```python
import random
from sklearn import preprocessing, metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics  import  accuracy_score , precision_score , recall_score,confusion_matrix,classification_report
```

```python
le = preprocessing.LabelEncoder()
y_le= le.fit(y)
y_trans = y_le.transform(y)
y_trans
```

```python
schema_features= list(X.columns)
for i in schema_features:
    X_le = le.fit(X[i])
    X[i] = X_le.transform(X[i])
```

```python
X_test = update_df_test_.copy()
```

```python
schema_features= list(X_test.columns)
for i in schema_features:
    X_le = le.fit(X_test[i])
    X_test[i] = X_le.transform(X_test[i])
```

<h4>Splitting train.csv data into train/val data</h4>

```python
X_train, X_val, y_train, y_val = train_test_split(X, y_trans, train_size=0.7, random_state=42)
X_train.shape, X_val.shape
```

<h2>Building Machine learning model</h2>

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=100,
                                  random_state=0)
 
# fit the regressor with x and y data
regressor.fit(X, y_trans)
```

```python
score = regressor.score(X_train, y_train)
print("R-squared:", score) 
```

<h3>Prediction with test.csv data</h3>

```python
y_pred = regressor.predict(X_test)
y_pred
```
